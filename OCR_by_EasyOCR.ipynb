**Pre-Processing the Input before OCR**

i.e. Used Noise Removal Techniques
& Extracted Text from Image 3 downloaded from Internet after applying some pre-processing Technique

Image 3 - https://drive.google.com/file/d/1jrgliWAp6TaxfL-DfJ7apjenf8Ic7Mpl/view?usp=sharing

```python
img_3 = cv2.imread("/content/drive/MyDrive/Captcha.png")
img_3 = cv2.resize(img_3,(400,400))
```

Firstly, Directly applying OCR on Images which doesn't give promising results

```python
cv2_imshow(img_3)
```

```python
result_base = reader.readtext(img_3)
```

```python
result_base
```

Secondly applying Background to Grey for getting Image in GreyScale, which also doesn't result in very promising results but better then earlier

```python
grey = cv2.cvtColor(img_3,cv2.COLOR_BGR2GRAY)
```

```python
cv2_imshow(grey)
```

```python
result_grey = reader.readtext(grey)
result_grey
```

Finally applying Grey Filter for approximately perfect result,

P.S. Selected Greyscale by Trial & Error

```python
grey_filtered = cv2.inRange(grey,100,105)
```

```python
cv2_imshow(grey_filtered)
```

```python
result_final = reader.readtext(grey_filtered)
result_final
```
